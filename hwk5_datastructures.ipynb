## Scenario: The Hot Dog Drama Network

Each user has a *dramatic storyline* during WeinerDash.

We will build a graph where:
- nodes = event types
- edges = transitions between events

BUT… we are going to interpret it like a **dramatic food storyline**:

- `login` → "enters the hot dog universe"
- `view_menu` → "browses questionable toppings"
- `search` → "panics and searches 'is ketchup illegal?'"
- `add_to_cart` → "commits to the hot dog"
- `checkout` → "achieves hot dog enlightenment"
- `logout` → "leaves, forever changed"

---

## Step 1:  Import and create a graph


````python

import networkx as nx

G = nx.Graph()
G
````


## Step 2 : Add nodes

These are the event types.


````python

events_types = ["login", "view_menu", "search", "add_to_cart", "checkout", "logout"]

G.add_nodes_from(events_types)
G.nodes()
````


## Step 3 — Add edges

We connect events that happen *one after another*.

This represents how users move through the system.


````python

# Build edges from the event stream
for i in range(len(events) - 1):
    current_user, current_event = events[i]  # we are just talking the events list of tuple and unpacking it:) 
    next_user, next_event = events[i+1]

    # Only connect events within the SAME user
    if current_user == next_user:
        G.add_edge(current_event, next_event)   

list(G.edges())
````


## Step 4 — Ask simple graph questions

### Q1: What events can happen after "login"?


````python
 list(G.neighbors("login"))
````


### Q2: Is there a path from "login" to "checkout"?

This tells us whether a user *can* reach checkout at all.


````python
nx.has_path(G, "login", "checkout")
````


### Q3: What is the shortest path from login to checkout?

This shows the **most efficient hot dog journey**.


````python
nx.shortest_path(G, "login", "checkout")
````


## Step 5 — Draw the graph

This helps visualize the chaos.


````python

import matplotlib.pyplot as plt

nx.draw(G, with_labels=True, node_size=2000, node_color="lightblue")
plt.title("Hot Dog Decision Graph")
plt.show()
````
